<h1>FILL NULL VALUES</h1>

```python
crimes['Location Description'] = crimes['Location Description'].astype(str).fillna('unaccounted')
crimes['Community Area'] = crimes['Community Area'].fillna(crimes['Community Area'].mode()[0])
crimes['X Coordinate'] = crimes['X Coordinate'].fillna(crimes['X Coordinate'].mode()[0])
crimes['Y Coordinate'] = crimes['Y Coordinate'].fillna(crimes['Y Coordinate'].mode()[0])
crimes['Latitude'] = crimes['Latitude'].fillna(crimes['Latitude'].mode()[0])
crimes['Longitude'] = crimes['Longitude'].fillna(crimes['Longitude'].mode()[0])
crimes['Location'] = crimes['Location'].astype(str).fillna('unaccounted')
```

```python
crimes.isnull().sum()
```

---

<h1>ADJUSTING DATA TYPES</h1>

```python
crimes.dtypes
```

```python
print(crimes.columns.tolist())
```

```python
crimes['Primary Type'] = crimes['Primary Type'].astype(str)
crimes['Description'] = crimes['Description'].astype(str)
crimes['Location Description'] = crimes['Location Description'].astype(str)
crimes['FBI Code'] = crimes['FBI Code'].astype(str)
crimes['Arrest'] = crimes['Arrest'].fillna(False).astype(bool)
crimes['Domestic'] = crimes['Domestic'].fillna(False).astype(bool)
crimes['Beat'] = pd.to_numeric(crimes['Beat'], errors='coerce')
crimes['District'] = pd.to_numeric(crimes['District'], errors='coerce')
crimes['Ward'] = pd.to_numeric(crimes['Ward'], errors='coerce')
crimes['Community Area'] = pd.to_numeric(crimes['Community Area'], errors='coerce')
crimes['Latitude'] = pd.to_numeric(crimes['Latitude'], errors='coerce')
crimes['Longitude'] = pd.to_numeric(crimes['Longitude'], errors='coerce')
crimes['X Coordinate'] = pd.to_numeric(crimes['X Coordinate'], errors='coerce')
crimes['Y Coordinate'] = pd.to_numeric(crimes['Y Coordinate'], errors='coerce')
crimes['Date'] = pd.to_datetime(crimes['Date'], errors='coerce')
crimes['Updated On'] = pd.to_datetime(crimes['Updated On'], errors='coerce')
crimes["Year"] = crimes["Date"].dt.year
crimes["Month"] = crimes["Date"].dt.month
crimes["Day"] = crimes["Date"].dt.day
crimes["Hour"] = crimes["Date"].dt.hour
crimes["Weekday"] = crimes["Date"].dt.day_name()
```

<h1>Q1. What are the top 10 most common primary crime types?</h1>

```python
crimes['Primary Type'].value_counts().head(10).plot(kind='bar', figsize=(8,5))
plt.title("Top 10 Primary Crime Types")
plt.xlabel("Crime Type")
plt.ylabel("Count")
plt.show()
```

<h1>Q2. How many crimes happened per year?</h1>

<h1>Q26. Cumulative monthly crimes over time</h1>

```python
monthly = crimes.groupby(pd.to_datetime(crimes['Date']).dt.to_period('M')).size().sort_index()
monthly_cum = monthly.cumsum()
plt.figure(figsize=(12,5))
monthly_cum.plot(marker='o')
plt.title("Cumulative Monthly Crimes Over Time")
plt.xlabel("Month-Year")
plt.ylabel("Cumulative Number of Crimes")
plt.xticks(rotation=45)
plt.grid(alpha=0.3)
plt.show()
```

<h2>Insight: The cumulative line shows total recorded crimes accumulating over time and highlights periods with faster accumulation. Sudden steep slopes indicate months/years with unusually high incident reporting.</h2>
