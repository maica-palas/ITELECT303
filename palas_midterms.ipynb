<h1>Converting Datasets to Dataframes</h1>

---

<h1>FILL NULL VALUES</h1>

```python
crimes['Location Description'] = crimes['Location Description'].astype(str).fillna('unaccounted')
crimes['Community Area'] = crimes['Community Area'].fillna(crimes['Community Area'].mode()[0])
crimes['X Coordinate'] = crimes['X Coordinate'].fillna(crimes['X Coordinate'].mode()[0])
crimes['Y Coordinate'] = crimes['Y Coordinate'].fillna(crimes['Y Coordinate'].mode()[0])
crimes['Latitude'] = crimes['Latitude'].fillna(crimes['Latitude'].mode()[0])
crimes['Longitude'] = crimes['Longitude'].fillna(crimes['Longitude'].mode()[0])
crimes['Location'] = crimes['Location'].astype(str).fillna('unaccounted')
```

```python
crimes.isnull().sum()
```

---

<h1>ADJUSTING DATA TYPES</h1>

```python
crimes.dtypes
```

```python
print(crimes.columns.tolist())
```

```python
crimes['Primary Type'] = crimes['Primary Type'].astype(str)
crimes['Description'] = crimes['Description'].astype(str)
crimes['Location Description'] = crimes['Location Description'].astype(str)
crimes['FBI Code'] = crimes['FBI Code'].astype(str)
crimes['Arrest'] = crimes['Arrest'].fillna(False).astype(bool)
crimes['Domestic'] = crimes['Domestic'].fillna(False).astype(bool)
crimes['Beat'] = pd.to_numeric(crimes['Beat'], errors='coerce')
crimes['District'] = pd.to_numeric(crimes['District'], errors='coerce')
crimes['Ward'] = pd.to_numeric(crimes['Ward'], errors='coerce')
crimes['Community Area'] = pd.to_numeric(crimes['Community Area'], errors='coerce')
crimes['Latitude'] = pd.to_numeric(crimes['Latitude'], errors='coerce')
crimes['Longitude'] = pd.to_numeric(crimes['Longitude'], errors='coerce')
crimes['X Coordinate'] = pd.to_numeric(crimes['X Coordinate'], errors='coerce')
crimes['Y Coordinate'] = pd.to_numeric(crimes['Y Coordinate'], errors='coerce')
crimes['Date'] = pd.to_datetime(crimes['Date'], errors='coerce')
crimes['Updated On'] = pd.to_datetime(crimes['Updated On'], errors='coerce')
crimes["Year"] = crimes["Date"].dt.year
crimes["Month"] = crimes["Date"].dt.month
crimes["Day"] = crimes["Date"].dt.day
crimes["Weekday"] = crimes["Date"].dt.day_name()
```

---

<h1>QUESTIONS</h1>

<h1>Q1. What are the top 10 most common primary crime types?</h1>

```python
crimes['Primary Type'].value_counts().head(10).plot(kind='bar', figsize=(8,5))
plt.title("Top 10 Primary Crime Types")
plt.xlabel("Crime Type")
plt.ylabel("Count")
plt.show()
```

<h2>Insight: Theft appears as the most common crime in Chicago. This suggests that property-related crimes are a major issue..</h2>

<h1>Q2. How many crimes happened per year?</h1>

```python
crimes['Year'].value_counts().sort_index().plot(kind='bar', figsize=(8,5))
plt.title("Crimes per Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

<h2>Insight: Theft and battery dominate reported crimes.
Targeting these types could significantly lower overall crime numbers.</h2>

<h1>Q3. How do crimes vary by month?</h1>

```python
monthly_crimes = crimes['Month'].value_counts().sort_index()

plt.figure(figsize=(10,5))
monthly_crimes.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Monthly Crime Counts in Chicago')
plt.xlabel('Month')
plt.ylabel('Number of Crimes')
plt.tight_layout()
plt.show()
```

<h2>Insight: Crimes rise during the summer months, particularly July and August.
This seasonal trend suggests the need for focused prevention during warmer periods.</h2>

<h1>Q4. What percentage of crimes resulted in arrest?</h1>

```python
arrest_counts = crimes['Arrest'].value_counts()

arrest_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightblue','lightgray'])
plt.title('Proportion of Crimes Leading to Arrest')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

<h2>Insight: A relatively small proportion of crimes result in arrests.
Boosting investigation and clearance rates could improve public safety outcomes.</h2>

<h1>Q5. Which police district records the most crimes?</h1>

```python
crimes['District'].value_counts().head(10).plot(kind='bar', figsize=(8,5), color='pink', edgecolor='black')
plt.title("Top Districts by Number of Crimes")
plt.xlabel("District")
plt.ylabel("Count")
plt.show()
```

<h2>Insight: Arrest rates have fluctuated, with some decline in recent years.
This may reflect changes in policing strategies or reporting practices.</h2>

<h1>Q6. Which location description has the highest crime occurrence?</h1>

```python
top_locations = crimes['Location Description'].value_counts().head(10)

plt.figure(figsize=(10,5))
top_locations.plot(kind='bar', color='slateblue', edgecolor='black')
plt.title('Top 10 Crime Locations')
plt.xlabel('Location')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

<h2>Insight: Streets, residences, and apartments dominate as crime scenes.
Prevention measures focusing on these areas can have a significant impact.</h2>

<h1>Q7. Which district reported the most crimes?</h1>

```python
top_districts = crimes['District'].value_counts().head(10)

plt.figure(figsize=(10,5))
top_districts.plot(kind='bar', color='mediumseagreen', edgecolor='black')
plt.title('Top 10 Districts by Crime Count')
plt.xlabel('District')
plt.ylabel('Number of Crimes')
plt.tight_layout()
plt.show()
```

<h2>Insight: A few districts consistently lead in reported crime counts.
Resource allocation could be prioritized for these high-crime districts.</h2>

<h1>Q8. Crimes trend by year<h1>

```python
crimes.groupby('Year').size().plot(marker='o', figsize=(8,5))
plt.title("Crime Trend Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Crimes")
plt.show()
```

<h2>Insight: The line chart shows how crime levels rise and fall across years. The marker o highlights the data points..</h2>

<h1>Q9. What are the top 10 locations of crimes?</h1>

```python
crimes['Location Description'].value_counts().head(10).plot(
    kind='bar', 
    figsize=(8,5), 
    color='skyblue', 
    edgecolor='black'
)
plt.title("Top 10 Crime Locations")
plt.xlabel("Location")
plt.ylabel("Count")
plt.show()
```

<h2>Insight: Streets are the most common crime location. This shows crimes often happen in public spaces.</h2>

<h1>Q10. Arrest vs No Arres</h1>

```python
arrest_counts = crimes['Arrest'].value_counts()
plt.figure(figsize=(7,6))
plt.pie(arrest_counts, labels=['No','Yes'], autopct='%1.1f%%', startangle=90)
plt.title("Arrest vs No Arrest")
plt.show()
```

<h2>Insight: Most incidents do not lead to immediate arrests, shown by the larger pie slice. This suggests many cases remain unresolved at first report.</h2>

<h1>Q11. Crimes per Year by Month</h1>

```python
monthly_by_year = crimes.groupby(["Year","Month"]).size().unstack(0)
monthly_by_year.plot(marker="o", figsize=(12,6))
plt.title("Monthly Crimes by Year")
plt.show()
```

<h2>Insight: The line chart shows that monthly crime counts follow a recurring seasonal pattern each year, with higher crime activity during the summer months and lower counts during winter. This suggests that weather and seasonal behavior play a role in crime trends.</h2>

<h1>Q12. Top 5 Primary Typess</h1>

```python
crimes["Primary Type"].value_counts().head(5).plot.pie(autopct="%1.1f%%", figsize=(6,6))
plt.title("Top 5 Primary Types")
plt.show()
```

<h2>Insight: Theft, battery, and criminal damage make up more than half of reported crimes.</h2>

<h1>Q13. Domestic vs Non-Domestic for Top Crimes</h1>

```python
top6 = crimes["Primary Type"].value_counts().head(6).index
pd.crosstab(crimes["Primary Type"], crimes["Domestic"]).loc[top6].plot(
    kind="bar", stacked=True, figsize=(10,6)
)
plt.title("Domestic vs Non-Domestic for Top Crimes")
plt.show()
```

<h2>Insight: Some crimes, like assault, have a higher share of domestic cases compared to theft or burglary.</h2>

<h1>Q14. Monthly Crimes by Year</h1>

```python
monthly_by_year = crimes.groupby(["Year","Month"]).size().unstack(0)
monthly_by_year.plot(marker="o", figsize=(12,6))
plt.title("Monthly Crimes by Year")
plt.show()
```

<h2>Insight: Different years follow similar seasonal crime trends, with peaks during warmer months.</h2>

<h1>Q15. Crimes per Day of Month</h1>

<h1>Q16. Monthly Crime Trend</h1>

<h2>Insight: The trend line shows ups and downs over time, possibly influenced by events like economic changes or police policies.</h2>

<h1>Q17. Top 10 FBI Codes</h1>

```python
crimes["FBI Code"].value_counts().head(10).plot(
    kind="barh", figsize=(9,5)
)
plt.title("Top 10 FBI Codes")
plt.xlabel("Count")
plt.ylabel("FBI Code")
plt.show()
```

<h2>Insight: FBI classification shows that property crimes dominate, but violent offenses also appear in the top categories.</h2>

<h1>Q18. Correlation Heatmap</h1>

```python
num_cols = ["Beat","District","Ward","Community Area","X Coordinate","Y Coordinate","Latitude","Longitude"]

sns.heatmap(crimes[num_cols].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

<h2>Insight: Location-based variables (coordinates, district, area) are strongly correlated, confirming spatial clustering of crimes.</h2>

<h1>Q20. Crimes per Year-Month</h1>

```python
pivot2 = crimes.pivot_table(index="Month", columns="Year", values="ID", aggfunc="count")

plt.figure(figsize=(12,6))
pivot2.plot(marker="o")
plt.title("Crimes per Month Across Years")
plt.xlabel("Month")
plt.ylabel("Number of Crimes")
plt.legend(title="Year")
plt.grid(True, alpha=0.3)
plt.show()
```

<h2>Insight: Summer months show consistently higher crime activity across different years.</h2>
